# Simulación de Autómata Celular en Blender

Este proyecto utiliza Blender, un software de modelado y animación 3D, para crear una simulación de un autómata celular en una dimensión (1D). El autómata celular se representa utilizando cubos en 3D que cambian de estado a lo largo del tiempo siguiendo un conjunto de reglas predefinidas.

4. Ejecuta el script en Blender haciendo clic en el botón "Run Script" en el editor de texto.

## Estructura del Código

El código proporcionado está diseñado para ejecutarse en Blender y utiliza la biblioteca NumPy para realizar cálculos. El autómata celular se modela como una matriz 2D de cubos en 3D que cambian de estado según las reglas definidas por `rule_number`.

```python
import bpy
import numpy as np

class CellularAutomata():
  def __init__(self, n_steps, rule_number):
    self.n_steps = n_steps
    self.rule_array = np.array([int(bit) for bit in np.binary_repr(rule_number, 8)])
    self.triplets = np.array([np.binary_repr(i, 3) for i in range(8)])
    self.rule_neigh ={key : val for key, val in zip(self.triplets[::-1], self.rule_array)}

  def fit(self, initial_state):
    self.m_cells = len(initial_state)
    self.CA_array = np.zeros((self.n_steps, self.m_cells))
    self.CA_array[0, :] = initial_state


    for step in range(1, self.n_steps):
      all_triplets = np.stack(
          [
              np.roll(self.CA_array[step - 1, :], 1),
              self.CA_array[step - 1, :],
              np.roll(self.CA_array[step - 1, :], -1),
          ]
      )
      gen_cell = []
      for i in range(self.m_cells):
          triplet = f'{int(all_triplets[0][i])}{int(all_triplets[1][i])}{int(all_triplets[2][i])}'
          gen_cell.append(self.rule_neigh[triplet])
      self.CA_array[step, :] = gen_cell

    return self.CA_array

rng = np.random.RandomState(42)
initial = rng.randint(0, 2, 50)
CA_1D = CellularAutomata(50, 90)

data = CA_1D.fit(initial)

pos_x = 0
for gen in data:
    pos_x += 2
    pos_y = 0
    for value in gen:
        pos_y += 2
        if value == 1:
            bpy.ops.mesh.primitive_cube_add(size=2, location=(pos_x+2, pos_y+2, 0), scale=(1, 1, 1))
```
